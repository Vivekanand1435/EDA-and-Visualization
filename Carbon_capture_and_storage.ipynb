```python
import numpy as np

# Simulate a dataset
np.random.seed(42)  # For reproducibility
data = {
    'Facility': [f'Facility {i}' for i in range(1, 101)],
    'Country': np.random.choice(['USA', 'Canada', 'Australia', 'China', 'Norway'], 100),
    'Capture Capacity (Mtpa)': np.random.uniform(0.5, 10, 100).round(2),
    'Storage Type': np.random.choice(['Saline Aquifer', 'Depleted Oil Field', 'Basalt Formation'], 100),
    'Status': np.random.choice(['Operational', 'Under Construction', 'Planned'], 100),
    'Year': np.random.randint(2000, 2025, 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the first few rows
print(df.head())
```

```python
# Duplicates: Have we identified and removed duplicate rows?
df.duplicated().sum()
```

```python
# Check dataset structure and data types
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Display summary statistics for numerical columns
print(df.describe())
```

```python
# Remove duplicates if any
df.drop_duplicates(inplace=True)

# Handling missing values (if applicable)
df.fillna({'Capture Capacity (Mtpa)': df['Capture Capacity (Mtpa)'].mean()
}, inplace=True)
```

```python
# Aggregate capture capacity by country
country_capacity = df.groupby('Country')['Capture Capacity (Mtpa)'].sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Capture Capacity (Mtpa)', data=country_capacity, palette='viridis')
plt.title('Total CO₂ Capture Capacity by Country')
plt.xlabel('Country')
plt.ylabel('Total Capture Capacity (Mtpa)')
plt.show()
```

```python
# Inference
# Countries like the USA and China dominate CO₂ capture capacity, reflecting advanced industrial CCS deployment, while others show limited adoption, highlighting opportunities for expansion in underrepresented regions.
```

```python
# Plot storage type distribution
plt.figure(figsize=(8, 6))
df['Storage Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Storage Types')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
# Inference
# Saline aquifers are the most utilized storage type, indicating their geological suitability and capacity for large-scale CCS, while basalt formations remain underexplored, suggesting potential for innovation.
```

```python
# Count the number of facilities by status
status_count = df['Status'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=status_count.index, y=status_count.values, palette='plasma')
plt.title('Facility Status Distribution')
plt.xlabel('Status')
plt.ylabel('Number of Facilities')
plt.show()
```

```python
# Inference
# A high proportion of planned facilities suggests rapid sector growth, but a smaller share of operational facilities highlights challenges in project completion or deployment readiness.
```

```python
# Aggregate capture capacity by year
yearly_capacity = df.groupby('Year')['Capture Capacity (Mtpa)'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Capture Capacity (Mtpa)', data=yearly_capacity, marker='o', color='blue')
plt.title('Trend of CO₂ Capture Capacity Over Time')
plt.xlabel('Year')
plt.ylabel('Total Capture Capacity (Mtpa)')
plt.grid(True)
plt.show()
```

```python
# Inference: CO₂ capture capacity exhibits consistent growth, with notable acceleration post-2015, likely driven by international climate agreements and increased funding for CCS technologies.
```

```python
# Create a pivot table for the heatmap
heatmap_data = df.pivot_table(values='Capture Capacity (Mtpa)', index='Country', columns='Storage Type', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('CO₂ Capture Capacity by Country and Storage Type')
plt.xlabel('Storage Type')
plt.ylabel('Country')
plt.show()
```

```python
# Inference: Countries exhibit distinct preferences for storage types—e.g., Norway leverages saline aquifers, while the USA employs a mix of depleted oil fields and saline aquifers—indicating regional specialization based on geological and industrial factors.
```
